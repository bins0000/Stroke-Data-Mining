```python
# create root window
root = Tk()
 
# root window title and dimension
root.title("Stroke Predictor Application")
# Set geometry (widthxheight)
root.geometry('700x700')

# adding a label to the root window
lbl = Label(root, text = "Are you a Geek?")
lbl.grid()

# adding Entry Field
txt = Entry(root, width=10)
txt.grid(column =1, row =0)

# function to display user text when
# button is clicked
def clicked():
 
    res = "You wrote" + txt.get()
    lbl.configure(text = res)

# button widget with red color text inside
btn = Button(root, text = "Click me" ,
             fg = "red", command=clicked)
# Set Button Grid
btn.grid(column=2, row=0)
    
# Execute Tkinter
root.mainloop()
```

```python
# import modules
from tkinter import *
from geopy.geocoders import Nominatim
from geopy import distance
 
# user defined function
def get_dis():
    try:
         
        geolocator = Nominatim(user_agent="geoapiExercises")
         
        place1 = geolocator.geocode(str(e1.get()))
        place2 = geolocator.geocode(str(e2.get()))
 
 
        Loc1_lat,Loc1_lon = (place1.latitude),(place1.longitude)
        Loc2_lat,Loc2_lon = (place2.latitude),(place2.longitude)
 
        location1=(Loc1_lat,Loc1_lon)
        location2=(Loc2_lat,Loc2_lon)
 
        res = (str(distance.distance(location1, location2).km)+" Km")
 
        result.set(res)
    except:
        result.set("something went wrong")

        
# object of tkinter
# with background set to light grey
master = Tk()
master.configure(bg='white')
master.title("Stroke Predictor Application")
 
# Variable Classes in tkinter
result = StringVar();
 

# Creating label for each information
# name using widget Label
Label(master, text="Enter gender (0: Male, 1: Female, 2: Others): " , bg = "white").grid(row=1, sticky=W)
Label(master, text="Enter age (year-old): " , bg = "white").grid(row=2, sticky=W)
Label(master, text="Enter hypertension (0: patient has hypertension, 1: patient has no hypertension): " , bg = "white").grid(row=3, sticky=W)

Label(master, text="Enter heart_disease (year-old): " , bg = "white").grid(row=4, sticky=W)
Label(master, text="Enter ever_married (year-old): " , bg = "white").grid(row=5, sticky=W)
Label(master, text="Enter work_type (year-old): " , bg = "white").grid(row=6, sticky=W)
Label(master, text="Enter Residence_type (year-old): " , bg = "white").grid(row=7, sticky=W)
Label(master, text="Enter avg_glucose_level (year-old): " , bg = "white").grid(row=8, sticky=W)
Label(master, text="Enter bmi (year-old): " , bg = "white").grid(row=9, sticky=W)
Label(master, text="Enter smoking_status (year-old): " , bg = "white").grid(row=10, sticky=W)

# label for the result
Label(master, text="Result :", bg = "white").grid(row=12, sticky=W)
 
# Creating label for class variable
# name using widget Entry
Label(master, text="", textvariable=result,bg = "white").grid(row=12,column=1, sticky=W)

# store variables
gender = Entry(master,width = 50)
gender.grid(row=1, column=1)
age = Entry(master,width = 50)
age.grid(row=2, column=1)
hypertension = Entry(master,width = 50)
hypertension.grid(row=3, column=1)

heart_disease = Entry(master,width = 50)
heart_disease.grid(row=4, column=1)
ever_married = Entry(master,width = 50)
ever_married.grid(row=5, column=1)
work_type = Entry(master,width = 50)
work_type.grid(row=6, column=1)
Residence_type = Entry(master,width = 50)
Residence_type.grid(row=7, column=1)
avg_glucose_level = Entry(master,width = 50)
avg_glucose_level.grid(row=8, column=1)
bmi = Entry(master,width = 50)
bmi.grid(row=9, column=1)
smoking_status = Entry(master,width = 50)
smoking_status.grid(row=10, column=1)

 
# creating a button using the widget 
b = Button(master, text="Predict", command=get_dis, bg = "white")
b.grid(row=11, column=1,columnspan=1, rowspan=1,padx=5, pady=5,)
 
mainloop()
```
